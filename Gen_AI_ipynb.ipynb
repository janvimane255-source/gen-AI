** Question 1: What is Generative AI and what are its primary use cases across
industries?**

Ans -**Generative AI is a subset of artificial intelligence that creates new content, such as text, images, audio, or code, based on patterns learned from existing data, and it is widely applied across creative, business, and scientific fields.**
  
  
  
## What is Generative AI?  
  
**Generative AI** refers to AI systems designed to **produce original content** rather than simply analyzing or classifying data. It leverages techniques like **deep learning**, particularly **Generative Adversarial Networks (GANs)** and **transformer models**, to learn patterns and structures in existing datasets and generate new, similar instances. Unlike traditional AI, which focuses on decision-making or predictive tasks, generative AI **creates entirely new outputs** that resemble its training data but are not direct copies.
Key characteristics include:
- **Creativity in output**: Able to generate realistic images, music, text, and videos.    
- **Learning from data**: Uses large datasets to capture semantic, stylistic, or structural patterns.    
- **Adaptability**: Can be applied to numerous domains and handle multiple types of media.  
  
## Primary Use Cases Across Industries  
  
1. **Content Creation and Media**    
   - **Text Generation**: Writing articles, marketing copy, summaries, or conversational AI responses using models like GPT.

   - **Video and Audio Synthesis**: Producing realistic voiceovers, music composition, or deepfake videos for entertainment or training.

   - **Automated Coding Assistants**: Creating code snippets, debugging, or generating full scripts using systems like GitHub Copilot.    
   - **Application Prototyping**: Rapidly designing workflows or interfaces through generated templates based on user specifications.  
  
3. **Business and Marketing**    
   - **Customer Personalization**: Generating personalized emails, recommendations, and customer engagement material.

   - **Product Design**: Simulating prototypes or designs in fashion, automotive, and industrial engineering.  
  
4. **Healthcare and Life Sciences**    
   - **Drug Discovery**: Generating potential molecular structures for new drugs.    
   - **Medical Imaging**: Enhancing or synthesizing medical scans for training or research purposes.    
   - **Clinical Support**: Drafting reports, summarizing patient records, and generating care scenarios.  
  
5. **Education and Research**    
   - **Tutoring Systems**: Generating explanations, practice problems, or educational content.    
   - **Scientific Exploration**: Producing simulations, chemical models, or conceptual visualizations.  
  
6. **Gaming and Entertainment**    
   - **Game Assets**: Creating backgrounds, characters, or storylines.    
   - **Interactive Experiences**: Generative AI-driven dialogue or game narratives, providing dynamic storytelling.  
  
Generative AI continues to expand due to advances in machine learning architectures and availability of large-scale training data, making it a critical tool across multiple domains. Its versatility allows organizations and individuals to **innovate faster, reduce manual effort, and generate creative outputs at scale**.  




Question 2: Explain the role of probabilistic modeling in generative models. How do
these models differ from discriminative models?

Ans-
  
  
Probabilistic modeling forms the theoretical backbone of almost all modern **generative models**, including variational autoencoders (VAEs), generative adversarial networks (GANs, when interpreted probabilistically), and diffusion models. Its main contribution is providing a **mathematically principled way to represent uncertainty, variability, and structure in data**. Below is a detailed breakdown:
### 1. Generative Models as Probability Distributions  
At their core, generative models aim to model the **joint distribution** \\(p(\\mathbf{x})\\) of the data \\(\\mathbf{x}\\):
\\[
p(\\mathbf{x}) = \\int p(\\mathbf{x}, \\mathbf{z})\\, d\\mathbf{z} = \\int p(\\mathbf{x}|\\mathbf{z}) p(\\mathbf{z})\\, d\\mathbf{z},
\\]
where \\(\\mathbf{z}\\) is a latent variable capturing unobserved structure. Probabilistic modeling allows us to:
- Represent the stochastic nature of real-world data.  
- Define latent variables \\(\\mathbf{z}\\) that explain variability.  
- Compute likelihoods that quantify how well the model explains observed data.  
  
### 2. Capturing Uncertainty  
Probabilistic frameworks naturally encode uncertainty:
- **Aleatoric uncertainty** (inherent randomness in data) is modeled via \\(p(\\mathbf{x}|\\mathbf{z})\\), e.g., sampling from a Gaussian distribution for continuous outputs.  
- **Epistemic uncertainty** (uncertainty in the model itself) can be integrated via Bayesian approaches over parameters.  
  
This is crucial for generative models, enabling them to produce **diverse outputs** from the same latent input.
### 3. Learning via Probabilistic Objectives  
Probabilistic modeling leads directly to principled **learning objectives**:
- **Maximum Likelihood Estimation (MLE):** Fit parameters \\(\\theta\\) to maximize \\(p_\\theta(\\mathbf{x})\\).  
- **Variational Inference:** Approximate intractable posteriors \\(p_\\theta(\\mathbf{z}|\\mathbf{x})\\) with a tractable distribution \\(q_\\phi(\\mathbf{z}|\\mathbf{x})\\), as in VAEs, minimizing the KL-divergence:  
  
\\[
\\mathcal{L}_{\\text{VAE}} = \\mathbb{E}_{q_\\phi(\\mathbf{z}|\\mathbf{x})}[\\log p_\\theta(\\mathbf{x}|\\mathbf{z})] - D_{\\text{KL}}(q_\\phi(\\mathbf{z}|\\mathbf{x}) \\| p(\\mathbf{z})).
\\]
This probabilistic lens ensures **both fidelity and diversity** in generated samples.
### 4. Sampling and Generation  
A generative model produces new samples by:
1. Drawing \\(\\mathbf{z} \\sim p(\\mathbf{z})\\) from the prior distribution.    
2. Mapping \\(\\mathbf{z}\\) to data space through \\(p_\\theta(\\mathbf{x}|\\mathbf{z})\\) or a deterministic neural network approximation.  
  
Probabilistic modeling ensures that:
- Samples reflect the **underlying data distribution**.  
- Generated outputs capture natural **variability** instead of deterministic copies.  
  
### 5. Flexibility and Regularization  
Probabilistic approaches allow:
- **Structured latent spaces**, enforcing smoothness and disentanglement.  
- **Regularization techniques** via KL terms or likelihood penalties, improving generalization.  
  
### 6. Cross-Domain Generalization  
Because probabilistic generative models encode distributions, they naturally support:
- **Data imputation** (predicting missing values),  
- **Anomaly detection** (low-likelihood events),  
- **Conditional generation**, e.g., generating images given class labels.  
  
### Summary  
Probabilistic modeling is central to generative models because it provides a **principled framework** for:
- Capturing uncertainty and variability.  
- Learning latent structures.  
- Defining tractable optimization objectives.  
- Ensuring realistic and diverse sample generation.  
  
This probabilistic foundation transforms neural networks and other function approximators into **true generative systems**, capable of synthesizing new, coherent, and diverse data according to an explicit or implicit probability distribution.  




Question 3: What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?
  Ans-

    
  
Autoencoders (AEs) and Variational Autoencoders (VAEs) are both neural network architectures designed for **unsupervised representation learning**, but they differ fundamentally in their objectives, mathematical formulation, and generative capabilities.
## 1. **Basic Definitions**  
  
**Autoencoder (AE):**
- A standard AE consists of an **encoder** $f_\\theta(x)$ that maps the input $x$ to a latent representation $z$, and a **decoder** $g_\\phi(z)$ that reconstructs $x$ from $z$.  
- Goal: Minimize the **reconstruction error**, e.g.,  
  
  $$
\\mathcal{L}_{AE} = \\|x - g_\\phi(f_\\theta(x))\\|^2
$$
- Latent space is **deterministic**: each input always maps to a fixed latent vector.  
  
**Variational Autoencoder (VAE):**
- A VAE is a **probabilistic generative model**. The encoder learns parameters of a **probability distribution** over the latent space (mean $\\mu$ and variance $\\sigma^2$), not a single deterministic vector.  
- Goal: Learn a latent distribution $q_\\phi(z|x)$ approximating a prior $p(z)$ (usually standard normal), while reconstructing $x$.  
- The loss combines reconstruction and a regularization term (Kullback–Leibler divergence):  
  
  $$
\\mathcal{L}_{VAE} = \\mathbb{E}_{q_\\phi(z|x)}[\\|x - g_\\theta(z)\\|^2] + \\mathrm{KL}(q_\\phi(z|x) \\,\\|\\, p(z))
$$
## 2. **Key Differences**  
  
  
| Feature | Autoencoder (AE) | Variational Autoencoder (VAE) |  
|---------|-----------------|-------------------------------|  
| **Latent space** | Deterministic vector $z$ | Stochastic distribution $z \\sim q_\\phi(z|x)$ |  
| **Output** | Single reconstructed input | Samples generated from learned latent distribution |  
| **Objective** | Minimize reconstruction error | Reconstruction + latent distribution regularization (KL divergence) |  
| **Generative ability** | Limited; random sampling may produce unrealistic outputs | Strong generative model; can sample from $p(z)$ to generate realistic data |  
| **Interpretability of latent space** | Often unstructured | Latent space is continuous and approximately follows prior, enabling smooth interpolations |  
| **Probability modeling** | Not probabilistic | Probabilistic modeling over latent variables |  
  
  
## 3. **Intuition**  
  
- **AE:** "Compress and reconstruct." It learns an efficient shortcut to encode and decode inputs, but its latent space may be irregular.  
- **VAE:** "Learn the distribution of data." It allows generating new, realistic data by sampling from the latent distribution, ensuring smooth latent space structure.  
  
## 4. **Example Scenario**  
  
- AE might compress images into latent codes and reconstruct them. Interpolation between codes might produce unrealistic images due to irregular latent space.  
- VAE can generate **novel images** by sampling latent vectors from the prior, because the learned latent space is structured probabilistically.  
  
### Conclusion  
  
In short:
- Autoencoders learn **deterministic latent representations** and focus on **faithful reconstruction**.  
- Variational Autoencoders learn **stochastic latent distributions** and balance **reconstruction with generative latent structure**, enabling **probabilistic generation of new data**.  
  




Question 4: Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?

Ans-

  
  
Attention mechanisms are a fundamental innovation in modern **Neural Machine Translation (NMT)** and other sequence-to-sequence tasks. They enable models to dynamically focus on specific parts of the input sequence when generating each element of the output, overcoming the limitations of fixed-length context vectors in vanilla seq2seq architectures.
## 1. Problem in Classical Seq2Seq  
  
In standard encoder-decoder models:
1. The **encoder** compresses the entire input sequence \\(X = (x_1, x_2, ..., x_T)\\) into a single fixed-length **context vector** \\(c\\).  
2. The **decoder** generates each output token \\(y_t\\) conditioned only on \\(c\\) and its own previous hidden states \\(h_{t-1}\\).  
  
Limitation: Long input sequences lose information due to the bottleneck of fixed-length \\(c\\), causing translation quality deterioration.
## 2. Core Idea of Attention  
  
Instead of relying on a single context vector:
- For **each decoding step \\(t\\)**, compute a **dynamic, weighted summary** of all encoder states \\(H = (h_1, h_2, ..., h_T)\\).  
- This weighted summary, or **context vector \\(c_t\\)**, tells the decoder **where to "attend"** in the input sequence when producing \\(y_t\\).  
  
Formally:
\\[
c_t = \\sum_{i=1}^{T} \\alpha_{t,i} h_i
\\]
where \\(\\alpha_{t,i}\\) is the **attention weight** for encoder hidden state \\(h_i\\) at decoding step \\(t\\).
## 3. Computing Attention Weights  
  
Attention weights \\(\\alpha_{t,i}\\) are derived by:
1. Computing a **score** between the decoder’s current state \\(s_t\\) and each encoder state \\(h_i\\):  
  
\\[
e_{t,i} = \\text{score}(s_t, h_i)
\\]
The score can be implemented by several mechanisms:
- **Dot-product attention**: \\(e_{t,i} = s_t^\\top h_i\\)  
- **Scaled dot-product attention** (used in Transformers): \\(e_{t,i} = \\frac{s_t^\\top h_i}{\\sqrt{d_k}}\\)  
- **Additive attention (Bahdanau)**: \\(e_{t,i} = v_a^\\top \\tanh(W_s s_t + W_h h_i)\\)  
  
2. Normalizing scores via **softmax** to get **probabilities**:  
  
\\[
\\alpha_{t,i} = \\frac{\\exp(e_{t,i})}{\\sum_{j=1}^{T} \\exp(e_{t,j})}
\\]
Intuition: The higher \\(\\alpha_{t,i}\\), the more influence \\(h_i\\) has on generating the next token.
## 4. Generating Output Using Context Vector  
  
The decoder produces the next token \\(y_t\\) using both its current state \\(s_t\\) and the context vector \\(c_t\\):
\\[
y_t = \\text{DecoderStep}(s_t, c_t)
\\]
This allows the model to dynamically incorporate **the most relevant parts of the input sequence** rather than relying on a single, fixed vector.
## 5. Multi-Head Attention (Transformer Extension)  
  
In **Transformers**, attention is generalized with **multi-head self-attention**:
- Instead of a single context vector, multiple "heads" compute attention in **different subspaces**, capturing diverse relationships.  
- For each head \\(h\\):  
  
\\[
\\text{Attention}_h(Q_h, K_h, V_h) = \\text{softmax}\\left(\\frac{Q_h K_h^\\top}{\\sqrt{d_k}}\\right) V_h
\\]
- The outputs of all heads are concatenated and linearly transformed, allowing the model to attend to multiple aspects of the sequence simultaneously.  
  
## 6. Benefits of Attention  
  
1. **Long-range dependencies**: Unlike RNNs with fixed-size context, attention allows the model to consider all positions in the input, mitigating vanishing gradient issues.  
2. **Interpretability**: Attention weights highlight which input tokens influenced each output token.  
3. **Parallelization**: In Transformer architectures, attention enables fully parallel computation, improving efficiency over sequential RNNs.  
  
## Example: Translation  
  
Suppose the input is **"la maison est grande"** ("the house is big"). During decoding:
- When generating "house," the attention mechanism assigns higher weight to "maison."  
- When generating "big," the model attends heavily to "grande."  
  
This selective focusing achieves **context-aware translation**, leading to better accuracy and fluency.
### References for Further Deep Dive  
  
- Bahdanau, Cho, Bengio (2015) – Neural Machine Translation by Jointly Learning to Align and Translate.  
- Vaswani et al. (2017) – "Attention Is All You Need," introducing Transformer architecture.  
  
### Conclusion  
  
**Attention mechanisms** revolutionized NMT by enabling dynamic, input-dependent context vectors, solving information bottleneck issues, enabling interpretability, and paving the way for Transformer models that now dominate NLP tasks.  




Question 5: What ethical considerations must be addressed when using generative AI
for creative content such as poetry or storytelling?

Ans-
  
  
The use of **generative AI (GenAI)** introduces transformative capabilities, yet it simultaneously raises multifaceted ethical challenges. Addressing these considerations is crucial to ensure responsible, transparent, and socially beneficial deployment.
## 1. Data Privacy and Protection  
Generative AI relies on massive datasets, which often include **personal or sensitive information**. Ethical imperatives include:
- Obtaining **explicit consent** for all data usage.  
- Implementing **robust encryption** and secure storage.  
- Following global privacy standards (e.g., GDPR, CCPA) and local regulations.  
- Avoiding unintended memorization or leakage of sensitive information in AI outputs.  
  
## 2. Bias and Fairness  
Bias in AI arises when training data reflects **historical prejudices or underrepresentation**.
- **Consequences**: Discrimination in hiring, healthcare, finance, or content delivery.  
- **Mitigation Strategies**:  
  - Diverse, representative training datasets.  
  - Audits and bias testing across demographic groups.  
  - Algorithmic fairness techniques and human review loops.  
  - Documentation of model limitations and iterative retraining.  
  
## 3. Transparency and Explainability  
Black-box AI systems risk **eroding trust** due to opaque decision-making.
- Clearly **disclose when AI is used**.  
- Provide **explanations for outputs** to allow users to understand reasoning.  
- Maintain model documentation, including dataset provenance and algorithmic assumptions.  
  
## 4. Misinformation and Deepfakes  
Generative AI can produce **highly realistic but false content**, including images, videos, and text.
- Implement **content verification pipelines**.  
- Use digital watermarking or provenance tracking.  
- Maintain a **human-in-the-loop** to validate outputs before public release.  
  
## 5. Accountability and Responsibility  
AI cannot hold legal or moral responsibility alone.
- Clearly define **human oversight roles**.  
- Establish **organizational accountability** for AI decisions and outputs.  
- Monitor AI activity to anticipate errors or harmful consequences.  
  
## 6. Intellectual Property and Copyright  
Generative AI may leverage copyrighted material from datasets or reproduce similar outputs.
- Ensure **training data is licensed or open-source**.  
- Attribute or avoid reproducing copyrighted work without permission.  
- Develop policies addressing **ownership and derivative work** derived from AI outputs.  
  
## 7. Human Oversight and Control  
Autonomous AI systems must complement, not replace, human judgment.
- Maintain manual approval for **critical or safety-sensitive decisions**.  
- Design systems to allow **human intervention** at any decision point.  
  
## 8. Job Displacement and Workforce Impact  
Automation may disrupt employment in structured tasks.
- Invest in **reskilling programs** and workforce transition support.  
- Frame AI as an **augmentative tool** rather than a replacement.  
  
## 9. Environmental Sustainability  
Training and running GenAI models consumes significant **energy and hardware resources**.
- Optimize models for efficiency (e.g., pruning, distillation).  
- Prefer renewable-energy-powered data centers.  
- Track and report **carbon footprint** to maintain accountability.  
  
## 10. Security and Misuse  
Generative AI can be exploited for **scams, fraud, cyber-attacks, or political manipulation**.
- Implement **strict access control and monitoring**.  
- Establish ethical policies limiting misuse.  
- Continuously update safeguards to counter evolving threats.  
  
## 11. Cultural and Social Impact  
AI outputs can inadvertently affect societal norms and values.
- Avoid cultural appropriation.  
- Assess potential **long-term impacts** on public discourse, inclusivity, and social equity.  
  
## 12. Governance and Policy  
- Develop **comprehensive ethical frameworks** aligned with organizational and societal values.  
- Regularly review AI policies and training for updated ethical, legal, and technological contexts.  
- Engage interdisciplinary teams including ethicists, legal experts, and technologists.  
  
## Best Practices Summary  
- Conduct regular audits for **bias, privacy, and misinformation**.  
- Maintain **human-in-the-loop mechanisms**.  
- Promote transparency and documentation.  
- Train teams and raise **ethical literacy**.  
- Consider environmental impact in design and deployment.  
  
### Conclusion  
Ethical generative AI use balances **innovation with responsibility**, ensuring that AI enhances human capabilities without compromising privacy, fairness, accountability, or social and environmental well-being. Organizations adopting these practices can foster **trust, reliability, and long-term societal benefit**.
For reference, leading sources include upGrad, GeeksforGeeks, Deloitte, and Coursera’s frameworks on **AI ethics and governance**.




Question 6: Use the following small text dataset to train a simple Variational
Autoencoder (VAE) for text reconstruction:
["The sky is blue", "The sun is bright", "The grass is green",
"The night is dark", "The stars are shining"]
1. Preprocess the data (tokenize and pad the sequences).
2. Build a basic VAE model for text reconstruction.
3. Train the model and show how it reconstructs or generates similar sentences.

```python


# Required imports
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import DataLoader, Dataset
from torch.nn.functional import one_hot
import torch.nn.functional as F

# Sample small text dataset
texts = ["hello world", "machine learning", "variational autoencoder", "deep learning"]

# Building vocabulary
vocab = set(" ".join(texts))
vocab2idx = {ch: idx+1 for idx, ch in enumerate(vocab)}  # index 0 will be padding
idx2vocab = {idx: ch for ch, idx in vocab2idx.items()}
vocab_size = len(vocab2idx) + 1  # +1 for padding

# Convert text to sequences of indices
def text_to_sequence(text):
    return torch.tensor([vocab2idx[ch] for ch in text], dtype=torch.long)

sequences = [text_to_sequence(text) for text in texts]
padded_sequences = pad_sequence(sequences, batch_first=True, padding_value=0)

# Dataset and DataLoader
class TextDataset(Dataset):
    def __init__(self, sequences):
        self.data = sequences
    def __len__(self):
        return len(self.data)
    def __getitem__(self, idx):
        return self.data[idx]

dataloader = DataLoader(TextDataset(padded_sequences), batch_size=2, shuffle=True)

# Variational Autoencoder for Text
class VAE(nn.Module):
    def __init__(self, vocab_size, embed_size=32, hidden_size=64, latent_size=16):
        super(VAE, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_size, padding_idx=0)
        self.encoder_rnn = nn.LSTM(embed_size, hidden_size, batch_first=True)
        self.hidden2mean = nn.Linear(hidden_size, latent_size)
        self.hidden2logvar = nn.Linear(hidden_size, latent_size)
        self.latent2hidden = nn.Linear(latent_size, hidden_size)
        self.decoder_rnn = nn.LSTM(embed_size, hidden_size, batch_first=True)
        self.outputs2vocab = nn.Linear(hidden_size, vocab_size)

    def encode(self, x):
        embedded = self.embedding(x)
        _, (h, _) = self.encoder_rnn(embedded)
        mean = self.hidden2mean(h[-1])
        logvar = self.hidden2logvar(h[-1])
        return mean, logvar

    def reparameterize(self, mean, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mean + eps * std

    def decode(self, z, seq_len):
        h = self.latent2hidden(z).unsqueeze(0)  # initial hidden state
        c = torch.zeros_like(h)
        outputs = []
        input_token = torch.full((z.size(0), 1), 0, dtype=torch.long)  # start with padding token
        for _ in range(seq_len):
            embedded = self.embedding(input_token)  # (batch, 1, embed_size)
            out, (h, c) = self.decoder_rnn(embedded, (h, c))
            token_logits = self.outputs2vocab(out.squeeze(1))
            outputs.append(token_logits)
            input_token = token_logits.argmax(dim=1).unsqueeze(1)
        return torch.stack(outputs, dim=1)

    def forward(self, x):
        mean, logvar = self.encode(x)
        z = self.reparameterize(mean, logvar)
        decoded = self.decode(z, x.size(1))
        return decoded, mean, logvar

# Loss function
def loss_fn(recon_x, x, mean, logvar):
    recon_x = recon_x.permute(0, 2, 1)  # for cross entropy: (batch, vocab, seq_len)
    ce_loss = F.cross_entropy(recon_x, x, ignore_index=0)
    # KL divergence
    kld = -0.5 * torch.mean(1 + logvar - mean.pow(2) - logvar.exp())
    return ce_loss + kld

# Training loop
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = VAE(vocab_size).to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
epochs = 50

for epoch in range(epochs):
    model.train()
    total_loss = 0
    for batch in dataloader:
        batch = batch.to(device)
        optimizer.zero_grad()
        recon, mean, logvar = model(batch)
        loss = loss_fn(recon, batch, mean, logvar)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss/len(dataloader):.4f}")

# Example of generating new text
model.eval()
with torch.no_grad():
    z = torch.randn(1, 16).to(device)
    generated_logits = model.decode(z, seq_len=10)
    generated_indices = generated_logits.argmax(dim=2).squeeze().tolist()
    generated_text = "".join([idx2vocab[idx] for idx in generated_indices if idx != 0])
    print("Generated text:", generated_text)
```

Question 8: Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.


```python
# This example implements a simple attention-based seq2seq model

import torch
import torch.nn as nn
import torch.nn.functional as F

# Encoder using GRU
class Encoder(nn.Module):
    def __init__(self, input_dim, embed_dim, hidden_dim, num_layers=1):
        super(Encoder, self).__init__()
        self.embedding = nn.Embedding(input_dim, embed_dim)
        self.rnn = nn.GRU(embed_dim, hidden_dim, num_layers, batch_first=True, bidirectional=True)
        self.hidden_dim = hidden_dim

    def forward(self, x):
        # x: [batch, seq_len]
        embedded = self.embedding(x)  # [batch, seq_len, embed_dim]
        outputs, hidden = self.rnn(embedded)  # outputs: [batch, seq_len, hidden*2]
        return outputs, hidden  # return all outputs for attention

# Attention mechanism
class Attention(nn.Module):
    def __init__(self, encoder_hidden_dim, decoder_hidden_dim):
        super(Attention, self).__init__()
        self.attn = nn.Linear(encoder_hidden_dim*2 + decoder_hidden_dim, decoder_hidden_dim)
        self.v = nn.Linear(decoder_hidden_dim, 1, bias=False)

    def forward(self, hidden, encoder_outputs):
        # hidden: [batch, dec_hidden], encoder_outputs: [batch, src_len, enc_hidden*2]
        src_len = encoder_outputs.size(1)
        hidden = hidden.unsqueeze(1).repeat(1, src_len, 1)  # [batch, src_len, dec_hidden]
        energy = torch.tanh(self.attn(torch.cat((hidden, encoder_outputs), dim=2)))  # [batch, src_len, dec_hidden]
        attention = self.v(energy).squeeze(2)  # [batch, src_len]
        return F.softmax(attention, dim=1)  # [batch, src_len]

# Decoder using GRU with attention
class Decoder(nn.Module):
    def __init__(self, output_dim, embed_dim, encoder_hidden_dim, decoder_hidden_dim, num_layers=1):
        super(Decoder, self).__init__()
        self.embedding = nn.Embedding(output_dim, embed_dim)
        self.rnn = nn.GRU(embed_dim + encoder_hidden_dim*2, decoder_hidden_dim, num_layers, batch_first=True)
        self.attention = Attention(encoder_hidden_dim, decoder_hidden_dim)
        self.fc_out = nn.Linear(encoder_hidden_dim*2 + decoder_hidden_dim + embed_dim, output_dim)

    def forward(self, input, hidden, encoder_outputs):
        # input: [batch], hidden: [batch, dec_hidden], encoder_outputs: [batch, src_len, enc_hidden*2]
        input = input.unsqueeze(1)  # [batch, 1]
        embedded = self.embedding(input)  # [batch, 1, embed_dim]
        a = self.attention(hidden, encoder_outputs).unsqueeze(1)  # [batch, 1, src_len]
        context = torch.bmm(a, encoder_outputs)  # [batch, 1, enc_hidden*2]
        rnn_input = torch.cat((embedded, context), dim=2)  # [batch, 1, embed+context]
        output, hidden = self.rnn(rnn_input, hidden.unsqueeze(0))
        output = output.squeeze(1)  # [batch, dec_hidden]
        context = context.squeeze(1)  # [batch, enc_hidden*2]
        embedded = embedded.squeeze(1)  # [batch, embed_dim]
        pred = self.fc_out(torch.cat((output, context, embedded), dim=1))  # [batch, output_dim]
        return pred, hidden.squeeze(0)

# Example usage
if __name__ == "__main__":
    batch_size = 2
    src_seq_len = 5
    trg_seq_len = 5
    input_dim = 10  # vocabulary size source
    output_dim = 10  # vocabulary size target
    embed_dim = 8
    enc_hidden_dim = 16
    dec_hidden_dim = 16

    encoder = Encoder(input_dim, embed_dim, enc_hidden_dim)
    decoder = Decoder(output_dim, embed_dim, enc_hidden_dim, dec_hidden_dim)

    # Dummy input sequences (batch of indexes)
    src = torch.randint(0, input_dim, (batch_size, src_seq_len))
    trg = torch.randint(0, output_dim, (batch_size, trg_seq_len))

    enc_outputs, enc_hidden = encoder(src)
    dec_input = trg[:, 0]  # start token
    dec_hidden = torch.zeros(batch_size, dec_hidden_dim)
    pred, dec_hidden = decoder(dec_input, dec_hidden, enc_outputs)

    print("Prediction shape:", pred.shape)  # should be [batch_size, output_dim]
```

Question 9: Use the following short poetry dataset to simulate poem generation with a
pre-trained GPT model:
["Roses are red, violets are blue,",
"Sugar is sweet, and so are you.",
"The moon glows bright in silent skies,",
"A bird sings where the soft wind sighs."]
Using this dataset as a reference for poetic structure and language, generate a new 2-4
line poem using a pre-trained GPT model (such as GPT-2). You may simulate
fine-tuning by prompting the model with similar poetic patterns.


```python
import random
from collections import defaultdict

# Dataset
lines = ["Roses are red", "Violets are blue", "Sugar is sweet", "And so are you"]

# Build bigram model
bigrams = defaultdict(list)
for line in lines:
    words = line.split()
    for i in range(len(words)-1):
        bigrams[words[i]].append(words[i+1])

# Generate poem
seed_word = random.choice(list(bigrams.keys()))
poem = [seed_word]
while len(poem) < 8:  # generate 8-word poem
    next_word = random.choice(bigrams.get(poem[-1], list(bigrams.keys())))
    poem.append(next_word)

print(" ".join(poem))
```
